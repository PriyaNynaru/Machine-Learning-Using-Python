###### Conditional Programming
- if
- elif
- else

```python
#write a program to check which number is greater
a=input("Enter the value ")
b=input("Enter the value  ")
if a>b:
    print(a,"is greater than",b)
else: 
    print(b,"is greater than",a)
```

```python
# Find Greatest of 3 numbers
a=input("Enter the value ")
b=input("Enter the value ")
c=input("Enter the value ")
if a>b and a>c:
    print(a)
elif b>c:
    print(b)
else:
    print(c)
```

##### Loops
    1. for
    2. while

```python
a=10
for i in range(a+1):
    print(i,end=" ")
```

```python
a=10
for i in range(1,a+1,5):
    print(i,end=" ")
```

```python
a=10
for i in range(a,0,-1):
    print(i,end=" ")
```

```python
a=10
for i in range(a,0,-3):
    print(i,end=" ")
```

```python
a="priya"
len(a)
```

```python
a="priya"
for i in range(len(a)): #prints index values
    print(i,end=" ")  
```

```python
a="priya"
for i in range(len(a)): #prints values
    print(a[i],end=" ")  
```

```python
a="string"
b="850053"
for i in a:
    for j in b:
        print(i,j,end=" ")
```

```python
a=1
while a<=10:
    print(a,end=" ")
    a+=1
```

```python
a=1
while a<=10:
    print(a,end=" ")
    if a==5:
        break
    a+=1
```

```python
a=1
while a<=10:
    print(a,end=" ")
    if a==5:
        continue
    a+=1
```
